```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("DATA.csv")
```

```python
print("\n🔹 REMOVING COMMAS FROM NUMERIC VALUES")

data = data.replace(',', '', regex=True)

print("Commas removed successfully!")
```

```python
print("\n🔹 FIXING DATA TYPE ISSUES (FORCE CLEANING)")

cols = ['OPEN','HIGH','LOW','PREV. CLOSE','LTP','CLOSE','VWAP','52W H','52W L','VALUE']

for col in cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

data['VOLUME'] = pd.to_numeric(data['VOLUME'], errors='coerce')
data['NO. OF  TRADES'] = pd.to_numeric(data['NO. OF  TRADES'], errors='coerce')

print("Conversion forced successfully!")
```

```python
print("\n🔹 CONVERTING DATE COLUMN")

data['DATE'] = pd.to_datetime(data['DATE'], format='%d-%b-%y')

print("Date column converted!")
print(data['DATE'].head())
print(data['DATE'].min(), data['DATE'].max())
print(data['DATE'].dtype)
```

```python
print("\n🔹 DROPPING UNUSED COLUMN")

data = data.drop(columns=['SERIES'])

print("SERIES column removed")
```

```python
print("\n🔹 SORTING DATA BY DATE")

data = data.sort_values(by='DATE')
data = data.reset_index(drop=True)

print("Data sorted successfully!")
data.head(3)
```

```python
print("\n🔹 UPDATED DATA TYPES")
print(data.dtypes)

print("\n🔹 STATISTICAL SUMMARY")
print(data.describe())
```

```python
data.to_csv("1cleaned_data.csv", index=False)
```
