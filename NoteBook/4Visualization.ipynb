```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
data = pd.read_csv("2feature_cleaned_data.csv")
```

```python


# 🔥 FORCE CLEAN DATETIME AGAIN (IMPORTANT)
data['DATE'] = pd.to_datetime(data['DATE'], errors='coerce')

plt.figure(figsize=(12,6))

plt.plot(data['DATE'], data['CLOSE'], label='Close Price')
plt.plot(data['DATE'], data['MA_20'], label='MA 20')
plt.plot(data['DATE'], data['MA_50'], label='MA 50')

plt.title("Stock Price Trend with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()

# 🔥 DO NOT USE ax.xaxis_date()

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python


plt.figure(figsize=(8,5))

sns.histplot(data['Return'], bins=50, kde=True)

plt.title("Distribution of Daily Returns")
plt.xlabel("Return")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(data['DATE'], data['Return'])

plt.title("Daily Returns Over Time")
plt.xlabel("Date")
plt.ylabel("Return")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(data['DATE'], data['VOLUME'])

plt.title("Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=data['Return'])

plt.title("Boxplot of Returns")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data['CLOSE'], bins=50, kde=True)

plt.title("Distribution of Closing Prices")

plt.show()
```

```python
data['MA_diff'] = data['MA_20'] - data['MA_50']

plt.figure(figsize=(12,5))

plt.plot(data['DATE'], data['MA_diff'])

plt.title("MA 20 - MA 50 Difference")
plt.xlabel("Date")
plt.ylabel("Difference")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x=data['Target'])
plt.title("Target Distribution")
plt.show()
```

```python
sns.boxplot(x=data['Target'], y=data['Return'])
plt.title("Return vs Target")
plt.show()
```
