```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("1cleaned_data.csv")
```

```python
print("\n🔹 CALCULATING DAILY RETURNS")

data['Return'] = data['CLOSE'].pct_change()

print(data[['DATE','CLOSE','Return']].head())
```

```python
print("\n🔹 CALCULATING MOVING AVERAGES")

data['MA_20'] = data['CLOSE'].rolling(window=20).mean()
data['MA_50'] = data['CLOSE'].rolling(window=50).mean()

print(data[['CLOSE','MA_20','MA_50']].head(10))
```

```python
print("\n🔹 CREATING TARGET VARIABLE")

data['Target'] = (data['CLOSE'].shift(-1) > data['CLOSE']).astype(int)

print(data[['CLOSE','Target']].tail())
```

```python
print("\n🔹 DROPPING NULL VALUES")

data = data.dropna()

print("New shape:", data.shape)
```

```python
data.to_csv("2feature_cleaned_data.csv", index=False)
```
